## <b>Motivação: Dados de clientes de um banco</b>

Considere uma amostra de 500 clientes de um banco no arquivo amostra_banco_distritos.csv. Estão disponíveis as variáveis:

- Cliente: Identificador do cliente.
- Sexo: Feminino (F) ou Masculino (M)
- Idade: Idade do cliente, em anos completos.
- Empresa: Tipo da empresa em que trabalha: Pública, Privada ou Autônomo
- Salário: Salário declarado pelo cliente na abertura da conta, em reais.
- Saldo_cc: Saldo em conta corrente, em reais.
- Saldo_poupança: Saldo em poupança, em reais.
- Saldo_investimento: Saldo em investimentos, em reais.
- Devedor_cartao: Valor em atraso no cartão de crédito, em reais.
- Inadimplente: Se o cliente é considerado inadimplente atualmente (1) ou não (0), de acordo com critérios preestabelecidos.
- Distrito: Distrito de São Paulo em que mora o cliente.


<b>Faça a leitura dos dados. Desenvolva a exploração e visualização dos dados. Verifique possíveis associações entre variáveis. Visualize dados em um mapa.</b>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leitura de dados

dados = pd.read_csv('https://raw.githubusercontent.com/cibelerusso/Datasets/main/amostra_banco_distrito.csv',sep=',', decimal=',')
dados.head(3)
```

### Agregação (agrupamento) de dados: groupby


```python
dados.groupby(by='Distrito').mean()
```

```python
# Exclui a coluna Cliente

dados.drop('Cliente', axis=1).groupby(by='Distrito').mean().reset_index()
```

### Associação entre variáveis quantitativas e qualitativas

```python
plt.rcParams["figure.figsize"] = [10, 15]
ax = sns.boxplot(y='Distrito', x='Salario', hue='Inadimplente',data=dados, palette='Set2')
```

```python
sns.pairplot(dados[['Salario','Saldo_cc', 'Saldo_poupança', 'Saldo_investimento', 'Devedor_cartao']])
```

```python
sns.pairplot(dados[['Salario','Saldo_cc', 'Saldo_poupança', 'Saldo_investimento', 'Devedor_cartao']], kind='reg')
```

```python
sns.pairplot(dados_nozeros[['Salario','Saldo_cc', 'Saldo_poupança', 'Saldo_investimento', 'Devedor_cartao']], kind='reg')
```

```python
dados_agregados = dados.drop('Cliente', axis=1).groupby(by='Distrito').mean()
dados_agregados
```

```python
# Dados de distritos com informações para mapa coroplético

import geopandas as gpd
#dados = pd.read_csv('https://raw.githubusercontent.com/cibelerusso/Datasets/main/amostra_banco_distrito.csv', decimal=',')
#dados_agregados = dados.drop('Cliente', axis=1).groupby(by='Distrito').mean()

distritos = gpd.read_file('https://raw.githubusercontent.com/cibelerusso/Datasets/8a162fdfc5c1f3bb8825cf8d87cdbff42790c21f/distritos-sp.geojson')
distritos.head(3)
```

### Merge

```python

plt.rcParams["figure.figsize"] = [15, 15]

all_data = pd.merge(dados_agregados, distritos, how='left', left_on='Distrito', right_on='ds_subpref')
all_data

gdf = gpd.GeoDataFrame(all_data, geometry='geometry')
gdf.plot(column='Inadimplente', legend=True, )
```

```python
from sklearn.preprocessing import StandardScaler
from scipy.cluster import hierarchy

df_scaled = dados_agregados.copy()
df_scaled.iloc[:, :] = StandardScaler().fit_transform(dados_agregados)

Z = hierarchy.linkage(df_scaled, 'ward')

plt.figure(figsize=(20, 10))
plt.grid(axis='y')
dn = hierarchy.dendrogram(Z, labels=list(df_scaled.index), leaf_font_size=10)
```

```python
df_scaled
```

```python
from sklearn.cluster import AgglomerativeClustering, KMeans
from scipy.cluster import hierarchy
```

```python
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters)
df_scaled['cluster'] = kmeans.fit_predict(df_scaled)
```

```python
plt.rcParams["figure.figsize"] = [15, 15]

all_data = pd.merge(df_scaled, distritos, how='left', left_on='Distrito', right_on='ds_subpref')
all_data

gdf = gpd.GeoDataFrame(all_data, geometry='geometry')
gdf.plot(column='cluster', legend=True, )
```
