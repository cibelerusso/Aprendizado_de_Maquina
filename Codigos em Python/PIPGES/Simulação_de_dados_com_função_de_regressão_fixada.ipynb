## Simulação de dados com função de regressão fixada

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fixa a semente para a simulação
np.random.seed(123)

# Criação valores para variável preditora x segundo uma grade igualmente espaçada
x = np.linspace(0, 10, 50)

# Cálculo dos valores de y usando a equação y = r(x) + ruído
# r(x) = a x^2 + b função de regressão linear nos parâmetros mas quadrática em X (teste outras)
def r(X, a, b):
    return a + b * X**2

# Geração de erro normal com média zero e variância 1000
erro = np.random.normal(0, np.sqrt(1000), len(x))

# Simulação dos y com a função de regressão fixada e os erros gerados
y = r(x,4,3) + erro
```

```python
# Gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Dados Simulados')
plt.plot(x, r(x,4,3), color='red', label='Função de regressão', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dados simulados segundo uma função de regressão fixada')
plt.legend()
plt.grid(True)
plt.show()
```
