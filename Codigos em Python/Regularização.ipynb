## LASSO

```python
# Adaptado de: https://machinelearningmastery.com/lasso-regression-with-python/

# Carrega as bibliotecas
from numpy import mean
from numpy import std
from numpy import absolute
from pandas import read_csv
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Leitura dos dados
url = 'https://raw.githubusercontent.com/cibelerusso/Aprendizado_de_Maquina/main/Dados/kc_house_data.csv'

df = read_csv(url)

X = df[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grade']].values
y = df['price'].values

# Definição do modelo
model = Lasso(alpha=10.0)

# Define o método de avalaição do modelo
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# Avalia o modelo
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)

# force scores to be positive
scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
# Carrega as bibliotecas
import pandas as pd
from numpy import arange
from sklearn.linear_model import LassoCV
from sklearn.model_selection import RepeatedKFold

# Leitura dos dados
url = 'https://raw.githubusercontent.com/cibelerusso/Aprendizado_de_Maquina/main/Dados/kc_house_data.csv'
df = pd.read_csv(url)

X = df[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grade']].values
y = df['price'].values

# Define o método de avalaição do modelo
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# Define o modelo
model = LassoCV(alphas=arange(0.1, 10, 0.1), cv=cv, n_jobs=-1)

# Ajusta o modelo
model.fit(X, y)

# summarize chosen configuration
print('alpha: %f' % model.alpha_)
```
