## Exemplo: MLP para função XOR (Ou exclusivo)


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

# Dados de treinamento XOR
X_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])

# Criar um classificador MLP
mlp = MLPClassifier(hidden_layer_sizes=(2,1), activation='logistic', solver='lbfgs', learning_rate_init=0.1, max_iter=1000, random_state=42)

# Treinar o classificador MLP
mlp.fit(X_train, y_xor)
```

```python
# Previsões para o conjunto de treinamento
predictions_xor = mlp.predict(X_train)
print("Previsões para XOR:", predictions_xor)
```

```python

# Plotar os dados de treinamento e as fronteiras de decisão
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_xor, cmap='viridis', edgecolors='k', label='Treinamento')
plt.scatter(X_train[:, 0], X_train[:, 1], c=predictions_xor, cmap='viridis', marker='s', edgecolors='k', label='Previsões')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(title='Classes')
plt.title('Dados de Treinamento e Fronteiras de Decisão - XOR')
plt.show()
```
